### Задача 1. Базовый уровень

На вход функция more_than_five(lst) получает список из целых чисел. 
Результатом работы функции должен стать новый список, в котором содержатся только те числа, которые больше 5 по модулю.

````python
import math

def more_than_five(lst):
    lst1 = []
    for i in lst:
        if abs(i) > 5:
            lst1.append(i)
    return lst1

a = []
for i in range(-10, 11):
    a.append(i)
print("Список ", a)
print("Подходящие числа: ", more_than_five(a))
````

Решить такое вручную, как вы понимаете, практически нереально. 
Вот Валентина и обратилась к вам за помощью. 
Помогите ей (при помощи функции all_divisors(number)). 
Постарайтесь найти самое оптимальное решение. 
Результат представьте в виде списка (не забудьте отсортировать по возрастанию).

````python
num1 = 23436
num2 = 190187200
num3 = 380457890232
def all_divisors(num):
    a = []
    for i in range(1, int((num + 1) ** 0.5)):
        if num % i == 0:
            a.append(i)
            b = int(num / i)
            if b not in a:
                a.append(b)
    print(sorted(a))
    
all_divisors(num1)
all_divisors(num2)
all_divisors(num3)
````

````python
import math

def all_divisors(number):
    """Возвращает список всех положительных делителей числа number."""
    divisors = set()  # Используем множество, чтобы избежать дубликатов
    for i in range(1, int(math.sqrt(number)) + 1):  # Проверяем до корня из number
        if number % i == 0:  # Если i — делитель number
            divisors.add(i)  # Добавляем i
            divisors.add(number // i)  # Добавляем парный делитель
    return sorted(divisors)  # Возвращаем отсортированный список

# Примеры использования:
numbers = [23436, 190187200, 380457890232]

for num in numbers:
    print(f"Делители числа {num}: {all_divisors(num)}")
````
